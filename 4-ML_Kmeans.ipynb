```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Select only numeric columns (excluding ZipCode, State names, city, etc.)
exclude_cols = ['ZipCode', 'State', 'USPS_ZIP_PREF_CITY', 'state_abbreviation']
model_numeric = model.select_dtypes(include=['float64', 'int64']).drop(columns=exclude_cols, errors='ignore')


# Step 2: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(model_numeric)

# Step 3: Run PCA (keep enough components to explain ~90% of variance)
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Step 4: Plot explained variance
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_.cumsum(), marker='o')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Select numeric columns only
exclude_cols = ['ZipCode', 'State', 'USPS_ZIP_PREF_CITY', 'state_abbreviation']
model_numeric = model.select_dtypes(include=['float64', 'int64']).drop(columns=exclude_cols, errors='ignore')


# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(model_numeric)

# Step 3: Apply PCA with 10 components
pca = PCA(n_components=10)
X_pca_10 = pca.fit_transform(X_scaled)

# Step 4: Add components to your model DataFrame
for i in range(10):
    model[f'PC{i+1}'] = X_pca_10[:, i]
```

```python
model
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Automatically detect all PC columns in your DataFrame
pca_columns = [col for col in model.columns if col.startswith('PC')]

# Just to confirm
print("PCA columns detected:", pca_columns)

# Create correlation matrix for PCA components
pca_corr = model[pca_columns].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pca_corr, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of PCA Components')
plt.tight_layout()
plt.show()
```

```python
features = model_numeric.columns.tolist()
loadings = pca.components_.T
num_pcs = loadings.shape[1]

loading_df = pd.DataFrame(loadings, columns=[f'PC{i+1}' for i in range(num_pcs)], index=features)

plt.figure(figsize=(12, 8))
sns.heatmap(loading_df, annot=True, cmap='coolwarm', center=0)
plt.title('PCA Component Loadings Heatmap', fontsize=16)
plt.xlabel('Principal Components', fontsize=14)
plt.ylabel('Original Features', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# Use only PC1 to PC10
pca_features = model[[f'PC{i+1}' for i in range(10)]]

# Elbow method to find optimal k
inertia = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_features)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Choose your k (e.g., from elbow plot â€” you can change it)
k_optimal = 10

# Run final KMeans
kmeans = KMeans(n_clusters=k_optimal, random_state=42, n_init = 20,max_iter=1000)
model['Cluster'] = kmeans.fit_predict(pca_features)
```
