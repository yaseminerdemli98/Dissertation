```python
rank = rank.sort_values('RankNetScore', ascending=False)
rank['rank_index'] = range(len(rank))
rank['rank_index'] = rank['rank_index'] + 1 
```

```python
ahp = pd.read_csv("/Users/yaseminerdemli/Documents/BA Courses/Dissertation/Project/Data/ahp_300_v1.csv")
ahp
```

```python
rank.columns
```

```python
rank = rank.drop(columns=['index','PC1','PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10'])
```

```python
ahp = ahp[['State', 'ZipCode', 'state_abbreviation', 'USPS_ZIP_PREF_CITY',
       'Higher Education', '% Unemployment', '% Business, Science, Arts',
       '% Poverty', 'Median Income', 'numberofStarbucks',
       'Average Daily Traffic Counts', 'Total Traffic Counts',
       'Count of Transit Stops', 'EmployeeNumber', 'near_city_center',
       'PotentialResidents', 'Competititors', 'HighIncomeDensity','TotalScore']]
```

```python
ahp = ahp.sort_values('TotalScore', ascending=False)
ahp['ahp_index'] = range(len(ahp))
ahp['ahp_index'] = ahp['ahp_index'] + 1 
```

```python
rank['rank']=1
```

```python
ahp['ahp']=1
```

```python
final = pd.merge(rank,
                 ahp, 
                 on = ['State', 'ZipCode', 'state_abbreviation', 'USPS_ZIP_PREF_CITY'],
                how = 'outer')
```

```python
final.columns
```

```python
import re

# find all the base names that have both an _x and an _y
cols = final.columns
bases = {
    re.sub(r'_[xy]$','',c)
    for c in cols
    if re.search(r'_[xy]$', c)
}

for base in bases:
    xcol = base + '_x'
    ycol = base + '_y'
    if xcol in cols and ycol in cols:
        final[base] = final[xcol].combine_first(final[ycol])
        final.drop([xcol, ycol], axis=1, inplace=True)
```

```python
import numpy as np

# Option A: nested np.where
final['common'] = np.where(
    (final['ahp'] == 1) & (final['rank'] == 1),      # both are 1
    'both',
    np.where(
        final['rank'] == 1,                         # else: rank is 1
        'rank',
        'ahp'                                       # else: ahp must be 1
    )
)
```
