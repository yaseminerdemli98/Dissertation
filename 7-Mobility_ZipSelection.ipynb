```python
def convert_and_format_zipcodes(df, dtype_map, errors='coerce'):
    """
    Converts data types, formats ZIP-code-like columns, and renames them to 'ZipCode'.

    Parameters:
    - df: pd.DataFrame
    - dtype_map: dict — column names and target data types (e.g., 'int', 'float', 'str', 'datetime')
    - errors: str — 'coerce', 'ignore', or 'raise' (default: 'coerce')

    Returns:
    - df: pd.DataFrame — with updated types and standardized ZIP code handling
    """
    for col, dtype in dtype_map.items():
        if col in df.columns:
            # If it's a ZIP-code-like column
            if 'zip' in col.lower():
                df[col] = df[col].dropna().astype(int).astype(str).str.zfill(5)
                df.rename(columns={col: 'ZipCode'}, inplace=True)
                col = 'ZipCode'  # Update reference for type conversion

            # Handle other type conversions
            if dtype == 'datetime':
                df[col] = pd.to_datetime(df[col], errors=errors)
            elif dtype in ['int', 'float']:
                df[col] = pd.to_numeric(df[col], errors=errors)
            elif dtype == 'str':
                df[col] = df[col].astype(str)
            else:
                df[col] = df[col].astype(dtype)
        else:
            print(f"⚠️ Column '{col}' not found in DataFrame.")
    return df
```

```python
dtype_map = {
    'ZIP CODE': 'int',
    'land_area': 'float',
    'people per sq mile': 'float',
    'Basic Education': 'float',              
    'Higher Education': 'float',            
    '% Income < 40k': 'float',
    '% 75k > Income > 40k': 'float',
    '% 125k > Income > 75k': 'float',
    '% Income > 125k': 'float',          
    '% Age < 18': 'float',            
    '% 29 > Age > 18': 'float',         
    '% 39 > Age > 30': 'float',         
    '% 49 > Age > 40': 'float',         
    '% 69 > Age > 50': 'float',         
    '% Age > 70': 'float',            
    '% Unemployment': 'float',         
    '% Business, Science, Arts': 'float',
    '% Poverty': 'float',         
    '% Affluence': 'float',          
    'Median Income': 'float'      
}

demog = convert_and_format_zipcodes(demographics, dtype_map)
```

```python
import numpy as np
import pandas as pd

# master = pd.DataFrame({
#     'State': [...],
#     'zipcode': [...],            # as strings or ints
#     'miles_to_zcta5': [...]
# })

# 1) Define conditions (most specific first)…
conditions = [
    # San Francisco Bay Area zips (start with '94' or '95')
    master['ZipCode'].astype(str).str.startswith(('94', '95')),

    # CA or TX (all other CA/TX)
    master['State'].isin(['California', 'Texas']),

    # DC, MD, VA
    master['State'].isin(['District of Columbia', 'Maryland', 'Virginia']),

    # New York
    master['State'] == 'New York',
]

# 2) …and their thresholds
thresholds = [
    2,    # Bay-Area ZIPs ≤ 2 miles
    3,    # CA & TX ≤ 3 miles
    1,    # DC/MD/VA ≤ 1 mile
    0.7,  # NY ≤ 0.7 miles
]

# 3) Attach the per-row threshold
master['state_threshold'] = np.select(conditions, thresholds, default=np.nan)

# 4) Filter
filtered = master[master['miles_to_zcta5'] <= master['state_threshold']].copy()

# 5) (Optional) clean up
filtered.drop(columns='state_threshold', inplace=True)
```

```python
filtered
```

```python
filtered = filtered[['ZipCode','zip2','miles_to_zcta5','State', 'state_abbreviation',
       'USPS_ZIP_PREF_CITY', 'PotentialResidents', 'Competititors','EmployeeNumber']]
filtered
```

```python
# 1) Make a weighted PR column:
filtered['PR_weighted'] = np.where(
    filtered['ZipCode'] == filtered['zip2'],
    filtered['PotentialResidents'],           # full value if they match
    filtered['PotentialResidents'] * 0.7      # otherwise 70%
)
```

```python
# 1) Make a weighted PR column:
filtered['Emp_weighted'] = np.where(
    filtered['ZipCode'] == filtered['zip2'],
    filtered['EmployeeNumber']*0.5,           # full value if they match
    filtered['EmployeeNumber'] * 0.25      # otherwise 70%
)
```

```python
#how many of the neighbor zipcodes will come
# 2) Group & sum using that weighted column
master_agg = (
    filtered
    .groupby('ZipCode', as_index=False)
    .agg(
        PotentialResidents = ('PR_weighted', 'sum'),
        EmployeeNumber     = ('Emp_weighted', 'sum'),
        ZipCount           = ('ZipCode', 'size'),
        IncludedZips       = ('zip2', lambda x: ','.join(sorted(map(str, x.unique()))))
    )
)
```

```python
master_agg['PotentialMember'] = (master_agg['PotentialResidents'] + 
                                master_agg['EmployeeNumber']) * 0.1
```

```python
master_agg['PotentialMember'] = master_agg['PotentialMember'].round()
```
