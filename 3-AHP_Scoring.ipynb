```python
# Again, pick the same columns to normalize:
cols_to_normalize = [
    'PotentialResidents',
    'HighIncomeDensity',
    'Average Daily Traffic Counts',
    'EmployeeNumber',
    'Median Household Income',
    'near_city_center',
    'Competititors' 
]

df_norm = master.copy()
for col in cols_to_normalize:
    # Convert column to numeric type first to handle any string values
    master[col] = pd.to_numeric(master[col], errors='coerce')
    
    # Compute square root of sum of squares
    denom = np.sqrt((master[col] ** 2).sum())
    df_norm[col + '_euclid'] = master[col] / denom
```

```python
# Step 1: Define pairwise comparisons
criteria_comparisons = {
    ('PotentialResidents', 'HighIncomeDensity'): 2,
    ('PotentialResidents', 'Average Daily Traffic Counts'): 7,
    ('PotentialResidents', 'EmployeeNumber'): 3,
    ('PotentialResidents', 'Median Household Income'): 3,
    ('PotentialResidents', 'near_city_center'): 5,
    ('PotentialResidents', 'Competititors'): 9,

    ('HighIncomeDensity', 'Average Daily Traffic Counts'): 5,
    ('HighIncomeDensity', 'EmployeeNumber'): 3,
    ('HighIncomeDensity', 'Median Household Income'): 4,
    ('HighIncomeDensity', 'near_city_center'): 4,
    ('HighIncomeDensity', 'Competititors'): 7,

    ('Average Daily Traffic Counts', 'EmployeeNumber'): 1/5,
    ('Average Daily Traffic Counts', 'Median Household Income'): 1/4,
    ('Average Daily Traffic Counts', 'near_city_center'): 1/2,
    ('Average Daily Traffic Counts', 'Competititors'): 1/2,

    ('EmployeeNumber', 'Median Household Income'): 3,
    ('EmployeeNumber', 'near_city_center'): 4,
    ('EmployeeNumber', 'Competititors'): 4,

    ('Median Household Income', 'near_city_center'): 2,
    ('Median Household Income', 'Competititors'): 2,

    ('near_city_center', 'Competititors'): 1/3,

    
}

# Step 2: Build the AHP Compare object
criteria = ahpy.Compare('Criteria', criteria_comparisons, precision=3)

# Step 3: Check priorities and consistency
criteria.report()
```

```python
df_norm['TotalScore'] = (df_norm['PotentialResidents_euclid'] * 0.347
                        + df_norm['HighIncomeDensity_euclid'] * 0.269
                        + df_norm['Average Daily Traffic Counts_euclid'] * 0.032
                        + df_norm['EmployeeNumber_euclid'] * 0.161
                        + df_norm['Median Household Income_euclid'] *  0.088
                        + df_norm['near_city_center_euclid'] * 0.047
                        + df_norm['Competititors_euclid'] * 0.056)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Define your criteria and corresponding weights
criteria_weights = {
    'Potential Residents': 0.347,
    'High Income Density': 0.269,
    'Employee Number': 0.161,
    'Median Household Income': 0.088,
    'Competitors': 0.056,
    'Near City Center': 0.047,
    'Average Daily Traffic Counts': 0.032
}

# Sort by weight descending
sorted_items = sorted(criteria_weights.items(), key=lambda x: x[1], reverse=True)
criteria, weights = zip(*sorted_items)

# Normalize weights for color mapping
norm_weights = np.array(weights)
norm_weights = (norm_weights - norm_weights.min()) / (norm_weights.max() - norm_weights.min())

# Use colormap
colors = cm.viridis(norm_weights)

# Plot
plt.figure(figsize=(8, 5))
bars = plt.barh(criteria, weights, color=colors)
plt.gca().invert_yaxis()  # Show highest weight at the top
plt.xlabel('Weight')
plt.title('AHP Criteria Weights')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Group and sort
state_counts = top300.groupby('State').size().sort_values(ascending=False)

# Normalize counts for color mapping
counts = state_counts.values
norm_counts = (counts - counts.min()) / (counts.max() - counts.min())
colors = cm.viridis(norm_counts)

# Plot
plt.figure(figsize=(10, 6))
bars = plt.barh(state_counts.index, counts, color=colors, edgecolor='black')
plt.gca().invert_yaxis()  # Highest count at top
plt.xlabel('Number of ZIP Codes in Top 300')
plt.title('Distribution of Top 300 ZIP Codes by State')
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height() / 2,
             str(int(width)),
             va='center', fontsize=9)

plt.tight_layout()
plt.show()
```
