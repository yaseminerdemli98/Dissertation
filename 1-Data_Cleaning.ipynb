```python

for city, item in cities.items():
    
    #df = pd.read_csv('Demographic data/Main Dataset.csv', dtype=str)
    df= pd.read_csv('/Users/yaseminerdemli/Documents/BA Courses/Dissertation/Project/Data/master_data_v2.csv', dtype=str,
                   delimiter = ',')
    # create column on number of people aged 35-55
    
    distances = pd.read_csv('/Users/yaseminerdemli/Documents/BA Courses/Dissertation/Project/Data/gaz2024zcta5distance50miles.csv', dtype=str,
                   delimiter = ',')
    
    #business_data = pd.read_csv('/Users/yaseminerdemli/Documents/BA Courses/Dissertation/Project/Data/business_data.csv', dtype=str,
                  # delimiter = ';')
    #business_data['Employee Number'] = business_data['Employee Number'].astype(float)
    #business_data = business_data.groupby('Zip Code')['Employee Number'].sum().reset_index()
    
    #df = df.merge(business_data, how='left', on = 'Zip Code')
    
    # filter to specific city
    df = df[df['ZIP CODE'].str.startswith(item[1])]
    
    df = df[['ZIP CODE', 'total_number_employees']]
    
    # add distances to other zips
    dist = distances.copy()
    dist = dist[dist['zip1'].str.startswith(item[1])]
    dist = dist.rename(columns = {'miles_to_zcta5': 'distance'})
    
    # filter to distances below 3 miles
    dist['distance'] = pd.to_numeric(dist['distance'], errors='coerce')
    dist = dist[dist['distance'] <= item[0]]
    
    # attach population to neighbouring zips
    neighbours = dist.copy()
    neighbours = neighbours.merge(df, how='left', left_on = 'zip2', right_on = 'ZIP CODE')
    neighbours = neighbours.rename(columns = {'total_number_employees': 'neighbouring_emp_catchment'})
    
    # group by zip code, sum the vlaues of the neighbouring areason rich people count
    emp_catchment = neighbours.groupby('zip1')['neighbouring_emp_catchment'].sum().reset_index()
    
    # join back on main data
    totals = df.merge(emp_catchment, how='left', left_on = 'ZIP CODE', right_on = 'zip1')

    ## NOT SURE !!!!
    totals.fillna({'total_number_employees': 0, 'neighbouring_emp_catchment': 0}, inplace=True)
    # Ensure both columns are numeric before multiplication
    totals['total_number_employees'] = pd.to_numeric(totals['total_number_employees'], errors='coerce').fillna(0)
    totals['neighbouring_emp_catchment'] = pd.to_numeric(totals['neighbouring_emp_catchment'], errors='coerce').fillna(0)
    
    # calculate score
    # score component 1 = 1 point for rich people in area
    # half a point for people in neighbouring areas
    totals.fillna({'total_number_employees': 0, 'neighbouring_emp_catchment': 0}, inplace=True)
    totals['emp_catchment_score'] = totals['total_number_employees'] + totals['neighbouring_emp_catchment'] * 0.5
    
    totals = totals[['ZIP CODE', 'total_number_employees',
                     'neighbouring_emp_catchment', 'emp_catchment_score']]
    
    totals['City'] = city
    
    commercial = pd.concat([commercial, totals], ignore_index=True)
    
    print(f'Done City: {city}')    
```
